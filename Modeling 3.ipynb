```python
import matplotlib.pyplot as plt

mu_values = [0.5, 1, 2, 5]
t_0 = 0
t_f = 100
n = 10000
initial_condition = [2, 0]
delta = (t_f - t_0) / n

def van_der_pol(t, y, mu):
    x, v = y
    dvdt = mu * (1 - x ** 2) * v - x
    return [v, dvdt]

def euler_method(mu):
    y = [initial_condition]
    t = t_0
    for _ in range(n):
        y_next = [y[-1][0] + delta * van_der_pol(t, y[-1], mu)[0],
                  y[-1][1] + delta * van_der_pol(t, y[-1], mu)[1]]
        t += delta
        y.append(y_next)
    return y

for mu in mu_values:
    y = euler_method(mu)
    x = [state[0] for state in y]
    v = [state[1] for state in y]
    plt.plot(x, v, label=f"μ = {mu}")

plt.xlabel('x')
plt.ylabel('dx/dt')
plt.title('Van der Pol Phase Plane Trajectories for Different μ (Euler\'s Method)')
plt.legend()
plt.show()
```
